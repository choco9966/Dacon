```python
import os, time
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook, tqdm
from selenium import webdriver as wd
from selenium.webdriver.common.by import By

tqdm.pandas()
```

```python
train = pd.read_csv('../data/train.csv', usecols = ['id','latitude','longitude'])
test  = pd.read_csv('../data/test.csv', usecols = ['id','latitude','longitude'])

combined = pd.concat([train, test],axis=0)

del train,test; 
```

```python
combined['latlong'] = combined['latitude'].astype(np.str) +'_'+combined['longitude'].astype(np.str)
```

```python
combined = combined.drop_duplicates(subset=['latlong']).reset_index(drop=True)
```

```python
combined.head()
```

```python
address_dict = {}
error_ids = []
for row in tqdm_notebook( combined[700:1400][['latlong','latitude','longitude']].iterrows() ):
    driver = wd.Chrome(executable_path = 'chromedriver')
    latlong, lat, long = row[1].values
    main_url = "https://www.google.com/maps/place/"
    driver.get(main_url + str(lat)+' '+ str(long))
    try:
        addresses = driver.find_elements(By.CLASS_NAME, value = 'section-info-text')
        address = [add.text for ind, add in enumerate(addresses) if ind == 1][0]
        print(latlong, address)
        address_dict[latlong] = address
    except:
        print(latlong)
        error_ids.append(latlong)
    driver.close()
    time.sleep(1)
```
